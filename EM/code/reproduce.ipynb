## EM 算法

### 自由能

$$F(q, \theta) = -E_q[logp(y, x|\theta)] - H(q)$$
$$ =-E_q[logp(y|x, \theta) + logp(x)] - H(q) $$

利用 **mean field approxiamation** 来估计后验分布 $p(x | y, \hat{\theta^{(t)}}) = h(x) = \Pi_ih_i(x_i)$

计算$ F_{MF}$

首先计算 $\log p(y_{i^{'}} | x_i^m, \theta) = [f(y^{k}_{i_{'}}; y_i^*, \Sigma)]^{v_i^{sk}}[g(y_{i^{'}}^k; h^k)]^{1-v_i^{sk}}$

```python
## 计算 ideal image 中 idx 的点在 kth image, 坐标为 x_ideal, homogeneous
## calibration = [K, R, T]
## 返回一个 homogenous coordinate
def map_ideal_onto_image(x_ideal, k, depth, calibration_ideal, calibration_k):
    K, R, T = calibration_k
    K_ideal, R_ideal, T_ideal = calibration_ideal
    tmp = depth * K @ R.T @ R_ideal @ np.linalg.inv(K_ideal) @ x_ideal + K @ R.T @ R_ideal @ T_ideal - K @ T
    return tmp / tmp[2]

## 计算第一部分
## @param: index 为ideal image 点的一维化坐标
# def part_1(index, )

## ideal_image 中的 index 为 idx 的点在 Kth image 上map的点为...的概率
## 映射至哪个点，通过给定的 calibration 可以算出来，因为我们算的概率是已知 x_i 的概率，
## 深度信息，calibration 信息已知，可以算出来它在任何一幅图中的 location, y_i^'




        
```

## util functions

```python
## 从 i, j 映射到一个一维数字 i * n_cols + j, i,j 从0开始

def homo(i, j):
    return [i, j, 1]

def dehomo(coordinate):
    return coordinate[:2]

def map_2D_to_1D(i, j, n_rows, n_cols):
    if j >= n_cols:
        print("width index out of range")
        return None
    if i >= n_rows:
        print("height index out of range")
        return None  
    return i * n_cols + j

def map_1D_to_2D(index, n_rows, n_cols):
    i = index // n_cols
    j = index - i * n_cols
    if i >= n_rows:
        print("height index out of range")
        return None
    if j >= n_cols:
        print("width index out of range")
        return None
    return (i, j)

## 从状态 m，映射到 Depth level r 和 visibility configuration s
## M = RS, m = (r - 1) * S + s, r, s 都从1开始
def map_m_to_r_s(m, R, S):
    if m % S == 0:
        r = m // S
    else:        
        r = m // S + 1
        
    s = m - (r - 1) * S
    if r > R:
        print(f"exceed max depth level: {R}")
    if s > S:
        print(f"exceed max visibility configuration level: {S}")
    
    return (r, s)

def map_r_s_to_m(r, s, R, S):
    if r > R:
        print(f"exceed max depth level: {R}")
    if s > S:
        print(f"exceed max visibility configuration level: {S}")
    return (r - 1) * S + s
```

## estimation

这一步要确定隐变量的分布，用 $b_i^m$ 来代表分布 ideal_image 中第 i 个点，是状态 m 的概率

计算 $\psi_{mn}$ $x_i, x_j, 在 状态 m, n的potential$
$\psi(x_{i}^{rs}, x_{j}^{pq}) = \exp({-D1 - D2}) + C$

$D1 = \dfrac{D_{ij}(r, p)}{\sigma_d}$

$D2 = \dfrac{D_{ij}(s, q)}{\sigma_v}$

```python
import math

## depth level 的距离
def cal_D1(r, p, sigma_d):
    D1 = math.abs(r - p) / (R * sigma_d)
    return D1

## visibility level 的距离
## number of dissimilar entries of v_i^s and v_j^q
## s, q configuration 的index，可以对应出两个向量
def cal_D2(s, q, sigma_v):
    s_conf = S_matrix[s]
    q_conf = S_matrix[q]
    dissimilar_count = 0 
    for i in range(len(s_conf)):
        if(s_conf[i] != q_conf[i]):
            dissimilar_count += 1            
    return dissimilar_count / sigma_v

## constant C

def psi_mn(m, n, C):
    r, s = map_m_to_r_s(m)
    p, q = map_m_to_r_s(n)
    return (math.exp(-cal_D1(r, p) - cal_D2(s, q)) + C)
```

## neighbor 信息

给定点 i(一维化坐标), 返回它的四个 neighbor

```python
def neighbor(index):
    i, j = map_1D_to_2D(index, height, width)
    left = map_2D_to_1D(i - 1, j, height, width)
    right = map_2D_to_1D(i + 1, j, height, width)
    upper = map_2D_to_1D(i, j + 1, height, width)
    lower = map_2D_to_1D(i, j - 1, height, width)
    return (left, right, upper, lower)
```

## 返回正态分布的概率

服从 $\mu = y_i^*, \Sigma$ 的正态分布

```python
def norm_pdf(x, mu, covariance):
    dim = len(x)
    return math.exp(-0.5 * (x - mu).T @ np.linalg.inv(covariance) @ (x - mu)) / ((2 * math.pi)^(dim / 2) * np.linalg.det(covariance)^0.5)
```

## 返回 histgram 的概率

先考虑灰度图 0 ~ 255

$h^k = \{h_b^k \}$

b 在 $[1, 255]$ 中取值，255个bins

$[0, 1] ... [254, 255]$

```python
def hist_prob(k, color):
    return h[k][int(color) + 1]
```

## E step

更新 $b_i^m$

```python
## 更新 bi_m
## given parameter and visible variable, calculate p(y_i^k)
## parameter 包括 ideal_image, 一个协方差矩阵，histogram
## kth 图片, k从0开始
## r depth level, s conf index
def update_bi_m_iterunit(index, m, k, r, s, covariance, calibration_ideal, calibration_k):
    ## 第一部分的概率
    conf_s = S_matrix[s]
    visible = conf_s[k]
    x_k_homo = map_ideal_onto_image(homo(map_1D_to_2D(index)), k, Depth[r], calibration_ideal, calibration_k)
    x_k = dehomo(x_k_homo)
    if visible:
        ## 正态分布, 其实是关于 rgb 三色的, 暂时定为灰度图
        color_prob =  norm_pdf(I[k][x_k], y_ideal[map_1D_to_2D(index)], covariance)
    else:
        ## 从 hist 中抽样
        color_prob = hist_prob(k, I[k][x_k])
        
    ## 第二部分的概率
    summation = 0
    for j in neighbor(index):
        for n in M:
            summation += b[j, n] * math.log(psi_mn(m, n, C))
    return color_prob * math.exp(summation)

def update_bi_m(i, m, r, s, covariance):
    res = 0
    for k in range(K):
        res += update_bi_m_iterunit(i, m, k, covariance, calibration_ideal, K_mat[k])
    return res
            
```
